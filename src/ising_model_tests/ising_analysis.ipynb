```python
import sys
sys.path.append("../")

from ising import Trajectory, filter_trajectories, magnetization, correlate
import numpy as np
from matplotlib import pyplot as plt
```

```python
Ns = [10, 25, 50, 100]

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

for N in Ns:

    filtered_trajectories = filter_trajectories(N=N, J=1, B=0, temperature=(0, 3.55), folder="ising_analysis_traj")

    temperatures = []
    magnetizations = []

    for name, params in filtered_trajectories.items():
        temperatures.append(params["temperature"])
        trajectory = Trajectory.from_file(name)
        magnetizations.append(trajectory.magnetization(r_equil=0.2, abs=True, n_blocks=1))

    indices = np.argsort(temperatures)

    ax.scatter(np.array(temperatures)[indices], np.array(magnetizations)[indices], label=f"N={N}", s=3)

n_1000 = np.loadtxt("ising_analysis_traj/out.csv", delimiter=",")
ax.scatter(n_1000[:, 0], n_1000[:, 1], label="N=1000", s=3)

ax.vlines(x=2.269185, ymin=0.0, ymax=1.1, ls="--", color="gray", label="$T_c = 2.269185$")

ax.set_title("Average magnetization per spin at different temperatures for different system sizes and $B = 0$")
ax.set_xlabel("Temperature (RU)")
ax.set_ylabel("$<|M|>/N$")

ax.legend()

plt.show()
```

```python
temperatures = [0.001, 0.3, 1.0, 2.0]

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

for temperature in temperatures:

    filtered_trajectories = filter_trajectories(N=10, J=0, B=(-2, 2), temperature=temperature, folder="ising_analysis_traj")

    Bs = []
    magnetizations = []

    for name, params in filtered_trajectories.items():
        Bs.append(params["B"])
        trajectory = Trajectory.from_file(name)
        magnetizations.append(trajectory.magnetization(r_equil=0.2, abs=False, n_blocks=1))

    indices = np.argsort(Bs)

    ax.scatter(np.array(Bs)[indices], np.array(magnetizations)[indices], label=f"T={temperature}", s=3)

    x = np.arange(-2, 2, 0.01)
    ax.plot(x, np.tanh(1/temperature * x), label=r"$\tanh(\beta B)$, $\beta = 1/" + str(temperature) + r"$")

ax.set_title("Average magnetization per spin at different magnetic field strengths with J=0 for different temperatures")
ax.set_xlabel("B")
ax.set_ylabel("$<|M|>/N$")

ax.legend()

plt.show()
```

```python
Ns = [10, 25, 50]

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

for N in Ns:

    filtered_trajectories = filter_trajectories(N=N, J=1, B=0, temperature=(0, 4.0), folder="ising_analysis_wolff_traj")

    temperatures = []
    magnetizations = []

    for name, params in filtered_trajectories.items():
        temperatures.append(params["temperature"])
        trajectory = Trajectory.from_file(name)
        magnetizations.append(trajectory.magnetization(r_equil=0.5, abs=True, n_blocks=1))

    indices = np.argsort(temperatures)

    ax.scatter(np.array(temperatures)[indices], np.array(magnetizations)[indices], label=f"N={N}", s=3)

ax.vlines(x=2.269185, ymin=0.0, ymax=1.1, ls="--", color="gray", label="$T_c = 2.269185$")

ax.set_title("Average magnetization per spin at different temperatures for different system sizes and $B = 0$ with Wolff algorithm")
ax.set_xlabel("Temperature (RU)")
ax.set_ylabel("$<|M|>/N$")

ax.legend()

plt.show()
```

```python
N = 10
temp = 4

traj = Trajectory.from_file(list(filter_trajectories(N=N, J=1, B=0, temperature=temp, folder="ising_analysis_wolff_traj").keys())[0])
mag = [np.abs(magnetization(configuration)) for configuration in traj.trajectory]

correlations = correlate(mag, mag)

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

ax.plot(correlations[:150])

ax.set_xlabel("k (lag time)")
ax.set_ylabel("$C_M(k) (Time autocorrelation function)$")

ax.set_title(f"Magnetization autocorrelation function for Wolff algorithm at N={N}, temp={temp}")

plt.show()
```

```python
N = 10
temperature = 0.8

fig, ax = plt.subplots(1, 1, figsize=(10, 7))

filtered_trajectories = filter_trajectories(N=N, J=1, B=0, temperature=temperature, folder="ising_analysis_wolff_traj")

for name, params in filtered_trajectories.items():
    magnetizations = []
    trajectory = Trajectory.from_file(name)
    for configuration in trajectory.trajectory:
        magnetizations.append(magnetization(configuration=configuration, normalize=True))

    ax.scatter(list(range(len(magnetizations))), magnetizations)

ax.set_title(f"Magnetization per spin at temperature {temperature} with $B = 0$ using Wolff algorithm")
ax.set_xlabel("Timestep")
ax.set_ylabel("$M/N$")

plt.show()
```
